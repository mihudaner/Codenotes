```python
%matplotlib inline
from concurrent.futures import thread
import enum
from tkinter.ttk import Sizegrip
import cv2
import matplotlib.pyplot as plt

#角点检测
img=cv2.imread(r"C:\Users\33567\Desktop\rect.png")
gray= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
dst=cv2.cornerHarris(gray,2,3,0.04)
img[dst>dst.max()*0.01]=[0,0,255]#列表生成器
cv2.imshow('ds',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
plt.figure(figsize=(15,15))
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.subplot(111),plt.imshow(img_rgb,'brg'),plt.title('img') 

```

```python
sift = cv2.xfeatures2d.SIFT_create()#特征检测的实例化
kp = sift.detect(gray, None)
img = cv2.drawKeypoints(gray, kp, img)
cv2.imshow('drawKeypoints', img)
cv2.waitKey(0)
cv2.destroyAllWindows()

kp, des = sift.compute(gray, kp)
plt.figure(figsize=(15,15))

plt.subplot(111),plt.imshow(img,'brg',),plt.title('img')
```
