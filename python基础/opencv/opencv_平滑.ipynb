```python
from ast import increment_lineno
from asyncio import constants
import cv2
from cv2 import cvtColor
from cv2 import threshold
from cv2 import THRESH_BINARY
from cv2 import THRESH_BINARY_INV
from cv2 import THRESH_TRUNC
from cv2 import THRESH_TOZERO
from cv2 import THRESH_TOZERO_INV
from cv2 import normalize#opencv读取格式BGR
import matplotlib
import matplotlib.pyplot as plt
import numpy as np#数值计算工具包
import threading
import matplotlib.pyplot as plt
def cv_show(name,img):
    cv2.imshow(name,img)
    cv2.waitKey(0)#waitKey应该是会影响其他进程，一直等待任意按键
    cv2.destroyAllWindows()

```

```python

#图像阈值方面函数
img_cat=cv2.imread(r'E:\codenotes\opencv\photo\img.png')
img_dog=cv2.imread(r'E:\codenotes\opencv\photo\img.png')
img_cat=cv2.cvtColor(img_cat,cv2.COLOR_BGR2GRAY)

ret,thresh1=cv2.threshold(img_cat,50,255,THRESH_BINARY)#二值图片
ret,thresh2=cv2.threshold(img_cat,127,255,THRESH_BINARY_INV)
ret,thresh3=cv2.threshold(img_cat,127,255,THRESH_TRUNC)#大于阈值的地方设为阈值，否则不变
ret,thresh4=cv2.threshold(img_cat,127,255,THRESH_TOZERO)#大于阈值的地方不变，小于阈值为0
ret,thresh5=cv2.threshold(img_cat,127,255,THRESH_TOZERO_INV)

titles=[1,2,3,4,5,6]
images=[img_cat,thresh1,thresh2,thresh3,thresh4,thresh5]
for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i]) 
    plt.xticks([]),plt.yticks([])
plt.show()

```

```python

#图像平滑处理

img_zaodian=cv2.imread(r'E:\codenotes\opencv\photo\img.png')
#img_zaodian=cv2.resize(img_zaodian,(0,0),fx=3,fy=3)
#均值滤波
img_zaodianed1=cv2.blur(img_zaodian,(3,3))
#图像方框滤波处理
img_zaodianed2=cv2.boxFilter(img_zaodian,-1,(3,3),normalize=True)#和均值结果一样
img_zaodianed3=cv2.boxFilter(img_zaodian,-1,(3,3),normalize=False)#和均值结果一样不会除以9
#高斯滤波
img_zaodianed4=cv2.GaussianBlur(img_zaodian,(5,5),1)
#中值滤波
img_zaodianed5=cv2.medianBlur(img_zaodian,7)
#print(img_zaodianed5)
res=np.hstack((img_zaodianed2,img_zaodianed4,img_zaodianed5))
'''水平（按列）顺序堆叠阵列。
函数concatenate、stack和block提供了更通用的堆叠和串联操作。'''
#print(res)
cv_show('ok',res)


plt.figure(figsize=(20,20))
plt.subplot(231),plt.imshow(img_zaodianed2,'brg'),plt.title('img_1') 
plt.subplot(232),plt.imshow(img_zaodianed4,'brg'),plt.title('mask') 
plt.subplot(233),plt.imshow(img_zaodianed5,'brg'),plt.title('mask_img') 





```
