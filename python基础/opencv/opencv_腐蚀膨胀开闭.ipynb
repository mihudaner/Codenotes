# 腐蚀膨胀


```python
from ast import increment_lineno
from asyncio import constants
import cv2
import matplotlib
import matplotlib.pyplot as plt
import numpy as np#数值计算工具包
import threading
import matplotlib.pyplot as plt

```

```python


#腐蚀操作
img_1=cv2.imread(r"E:\Work\THU\code\Zivid\temp\deepmap\000024_deepmap.jpg")

cv2.namedWindow('img', 0)# 窗口名字
cv2.resizeWindow('img', (500, 500))# 设置窗口大小
cv2.moveWindow("img", 50, 50)	# 显示框位置，左上角为原点（0， 0）坐标，第一个是x坐标，第二个是y坐标
cv2.imshow('img',img_1)
cv2.waitKey(0)
cv2.destroyAllWindows()

kernel=np.ones((5,5),np.uint8)
erosion1=cv2.erode(img_1,kernel,iterations=1)
erosion2=cv2.erode(img_1,kernel,iterations=2)
erosion3=cv2.erode(img_1,kernel,iterations=3)
res=np.hstack((erosion1,erosion2,erosion3))
cv2.imshow('img',res)#黑色部分会变粗
cv2.waitKey(0)
cv2.destroyAllWindows()

plt.subplot(231),plt.imshow(img_1,'brg'),plt.title('img_1') 
plt.subplot(232),plt.imshow(erosion1,'brg'),plt.title('erosion1') 
plt.subplot(233),plt.imshow(erosion2,'brg'),plt.title('erosion2') 


plt.subplot(235),plt.imshow(erosion3,'brg'),plt.title('erosion3') 
plt.subplot(236),plt.imshow(res,'brg'),plt.title('res') 
```

```python

#膨胀操作
dialate1=cv2.dilate(img_1,kernel,iterations=1)
dialate2=cv2.dilate(img_1,kernel,iterations=2)
dialate3=cv2.dilate(img_1,kernel,iterations=3)
res=np.hstack((dialate1,dialate2,dialate3))
cv2.imshow('img',res)#黑色部分会变粗
cv2.waitKey(0)
cv2.destroyAllWindows()
plt.subplot(231),plt.imshow(dialate1,'brg'),plt.title('dialate1') 
plt.subplot(232),plt.imshow(dialate2,'brg'),plt.title('dialate2') 
plt.subplot(233),plt.imshow(dialate3,'brg'),plt.title('dialate3') 
```

```python

#开运算，（先腐蚀再膨胀）
opening=cv2.morphologyEx(img_1,cv2.MORPH_OPEN,kernel)
cv2.imshow('img',opening)#黑色部分会变粗
cv2.waitKey(0)
cv2.destroyAllWindows()

#闭运算（先膨胀再腐蚀）
closing=cv2.morphologyEx(img_1,cv2.MORPH_CLOSE,kernel)
cv2.imshow('img',closing)#黑色部分会变粗
cv2.waitKey(0)
cv2.destroyAllWindows()

plt.subplot(231),plt.imshow(img_1,'brg'),plt.title('img_1') 
plt.subplot(232),plt.imshow(opening,'brg'),plt.title('opening') 
plt.subplot(233),plt.imshow(closing,'brg'),plt.title('closing') 
```
