```python
from ast import increment_lineno
from asyncio import constants
import cv2
from cv2 import cvtColor
from cv2 import threshold
from cv2 import THRESH_BINARY
from cv2 import THRESH_BINARY_INV
from cv2 import THRESH_TRUNC
from cv2 import THRESH_TOZERO
from cv2 import THRESH_TOZERO_INV
from cv2 import normalize
from cv2 import imread#opencv读取格式BGR
import matplotlib
import matplotlib.pyplot as plt
import numpy as np#数值计算工具包
import threading
import matplotlib.pyplot as plt
```

```python


#可以同时打开俩个窗口
img_1=cv2.imread(r'E:\project\python\data\1.jpg')
img_2=cv2.imread(r'E:\project\python\data\team.png')
img_2=cv2.resize(img_2,(800,600))
cv2.imshow('img1',img_1)
cv2.imshow('img2',img_2)
cv2.waitKey(1000)
cv2.destroyAllWindows()

#梯度  膨胀-腐蚀
kernel=np.ones((4,4),np.uint8)
erosion=cv2.erode(img_1,kernel,iterations=3)
dilate =cv2.dilate(img_1,kernel,iterations=3)
gradient1=(dilate-erosion)
gradient2=cv2.morphologyEx(img_1,cv2.MORPH_GRADIENT,kernel)
res=np.hstack((erosion,dilate,gradient1,gradient2))
cv2.imshow('img',res)
cv2.waitKey(1000)
cv2.destroyAllWindows()
plt.subplot(2,3,1),plt.imshow(img_1,'brg'),plt.title('img_1') 
plt.subplot(232),plt.imshow(img_2),plt.title('img_2') 
plt.subplot(233),plt.imshow(erosion,'brg'),plt.title('erosion') 
plt.subplot(234),plt.imshow(dilate,'brg'),plt.title('dilate') 
plt.subplot(235),plt.imshow(gradient1,'brg'),plt.title('gradient1') 
plt.subplot(236),plt.imshow(gradient2,'brg'),plt.title('gradient2') 
```

```python

#礼帽 原始输入-开运算结果
#黑帽 闭运算-原始输入
res1=cv2.morphologyEx(img_1,cv2.MORPH_TOPHAT,kernel)
res2=cv2.morphologyEx(img_1,cv2.MORPH_BLACKHAT,kernel)
res=np.hstack((res1,res2))
cv2.imshow('img',res)
cv2.waitKey(0)
cv2.destroyAllWindows()

plt.subplot(231),plt.imshow(img_1,'brg'),plt.title('img_1') 
plt.subplot(232),plt.imshow(res1,'brg'),plt.title('res1') 
plt.subplot(233),plt.imshow(res2,'brg'),plt.title('res2') 
```
