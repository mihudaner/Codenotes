```python
from ast import increment_lineno
from asyncio import constants
import cv2
from cv2 import cvtColor#opencv读取格式BGR
import matplotlib
import matplotlib.pyplot as plt
import numpy as np#数值计算工具包

import matplotlib.pyplot as plt

def cv_show(name,img):
    cv2.imshow(name,img)
    cv2.waitKey(1000)#waitKey应该是会影响其他进程，一直等待任意按键
    cv2.destroyAllWindows()
```

```python

#图像的边缘填充
img_cat=cv2.imread(r'C:\Users\wangkai\Desktop\pythonwork\opencv\photo\cat.jpg')
img_dog=cv2.imread(r'C:\Users\wangkai\Desktop\pythonwork\opencv\photo\dog.jpg')

top_size,bottom_size,left_size,right_size=(50,50,50,50)

replicate = cv2.copyMakeBorder(img_cat,top_size,bottom_size,left_size,right_size,borderType=cv2.BORDER_REPLICATE)
reflect = cv2.copyMakeBorder(img_cat,top_size,bottom_size,left_size,right_size,borderType=cv2.BORDER_REFLECT)
reflect101 = cv2.copyMakeBorder(img_cat,top_size,bottom_size,left_size,right_size,borderType=cv2.BORDER_REFLECT_101)
wrap = cv2.copyMakeBorder(img_cat,top_size,bottom_size,left_size,right_size,borderType=cv2.BORDER_CONSTANT,value=0)
constant = cv2.copyMakeBorder(img_cat,top_size,bottom_size,left_size,right_size,borderType=cv2.BORDER_CONSTANT,value=0)
plt.figure(figsize=(20,10))
plt.subplot(231),plt.imshow(img_cat,'brg'),plt.title('ORIGINAL') 
plt.subplot(232),plt.imshow(replicate,'gray'),plt.title('replicate') 
plt.subplot(233),plt.imshow(reflect,'gray'),plt.title('reflect') 
plt.subplot(234),plt.imshow(reflect101,'gray'),plt.title('reflect101') 
plt.subplot(235),plt.imshow(wrap,'gray'),plt.title('wrap') 
plt.subplot(236),plt.imshow(constant,'gray'),plt.title('constant') 
plt.show()
```

```python


#图像加法
img_cat1=img_cat+10
img_cat2=img_cat1+img_cat#直接相加，numpy的数值超过256会直接%256
img_cat3=cv2.add(img_cat1,img_cat2)#超过256等于255
plt.figure(figsize=(20,20))
plt.subplot(231),plt.imshow(img_cat1,'brg'),plt.title('cat1') 
plt.subplot(232),plt.imshow(img_cat2,'brg'),plt.title('cat2') 
plt.subplot(233),plt.imshow(img_cat3,'brg'),plt.title('cat3') 


#图像融合
img_cat4=cv2.resize(img_cat,(500,300))
img_dog4=cv2.resize(img_dog,(500,300))
img_cat_bigger=cv2.resize(img_cat,(0,0),fx=10,fy=10)#倍数
#cv2.imwrite()

img_merge=cv2.addWeighted(img_cat4,0.4,img_dog4,0.6,0)#权重微调
plt.figure(figsize=(20,20))
plt.subplot(234),plt.imshow(img_cat4,'brg'),plt.title('cat3') 
plt.subplot(235),plt.imshow(img_dog4,'brg'),plt.title('cat3') 
plt.subplot(236),plt.imshow(img_merge,'brg'),plt.title('cat3') 
print(img_cat4.shape)

plt.show()



```
