```python
import cv2
import numpy as np#数值计算工具包
import matplotlib.pyplot as plt
```

```python
img_1=cv2.imread(r"E:\codenotes\opencv\photo\img.png")


def cv_show(name,img):
    cv2.imshow(name,img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

cv_show('soblexy',img_1)
soblex=cv2.Sobel(img_1,cv2.CV_64F,0,1,ksize=3)#soblex内含有负数，要绝对值一下
soblex = cv2.convertScaleAbs(soblex)
sobley=cv2.Sobel(img_1,cv2.CV_64F,1,0,ksize=3)
sobley = cv2.convertScaleAbs(sobley)
soblexy=cv2.addWeighted(soblex,0.5,sobley,0.5,0)#用权值相加效果好一些
soblexy = cv2.convertScaleAbs(soblexy)
cv_show('soblexy',soblexy)

scharrx=cv2.Scharr(img_1,cv2.CV_64F,0,1)#soblex内含有负数，要绝对值一下
scharrx = cv2.convertScaleAbs(scharrx)
scharry=cv2.Scharr(img_1,cv2.CV_64F,1,0)
scharry = cv2.convertScaleAbs(sobley)
scharrxy=cv2.addWeighted(scharrx,0.5,scharry,0.5,0)#用权值相加效果好一些
scharrxy = cv2.convertScaleAbs(scharrxy)
cv_show('scharrxy',scharrxy)

laplacianx=cv2.Laplacian(img_1,cv2.CV_64F)#soblex内含有负数，要绝对值一下
laplacianx = cv2.convertScaleAbs(laplacianx)
laplaciany=cv2.Laplacian(img_1,cv2.CV_64F)
laplaciany = cv2.convertScaleAbs(laplaciany)
laplacianxy=cv2.addWeighted(laplacianx,0.5,laplaciany,0.5,0)#用权值相加效果好一些
laplacianxy = cv2.convertScaleAbs(laplacianxy)
cv_show('laplacianxy',laplacianxy)

```

```python

#图像梯度-Sobal算子
'''
[-1 0  1
-2  0  2
-1  0  1]
'''
img_1=cv2.imread(r"E:\codenotes\opencv\photo\img.png")
def cv_show(name,img):
    cv2.imshow(name,img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

soblex=cv2.Sobel(img_1,cv2.CV_64F,0,1,ksize=3)#soblex内含有负数，要绝对值一下
soblex = cv2.convertScaleAbs(soblex)
sobley=cv2.Sobel(img_1,cv2.CV_64F,1,0,ksize=3)
sobley = cv2.convertScaleAbs(sobley)
soble=cv2.Sobel(img_1,cv2.CV_64F,1,1,ksize=3)
soble = cv2.convertScaleAbs(soble)
soblexy=cv2.addWeighted(soblex,0.5,sobley,0.5,0)#用权值相加效果好一些
soblexy = cv2.convertScaleAbs(soblexy)

res=np.hstack((soblex,sobley,soble,soblexy))

cv_show('soblex',res)
plt.figure(figsize=(20,20))
plt.subplot(231),plt.imshow(soblex,'brg'),plt.title('soblex') 
plt.subplot(232),plt.imshow(sobley,'brg'),plt.title('sobley') 
plt.subplot(233),plt.imshow(soblexy,'brg'),plt.title('soblexy')

```

```python


#图像梯度-scharr算子
'''
[-3 0  3
-10 0  10
-3  0  3]
'''
scharrx=cv2.Scharr(img_1,cv2.CV_64F,0,1)#soblex内含有负数，要绝对值一下
scharrx = cv2.convertScaleAbs(scharrx)
scharry=cv2.Scharr(img_1,cv2.CV_64F,1,0)
scharry = cv2.convertScaleAbs(sobley)
scharrxy=cv2.addWeighted(scharrx,0.5,scharry,0.5,0)#用权值相加效果好一些
scharrxy = cv2.convertScaleAbs(scharrxy)
plt.figure(figsize=(20,20))
plt.subplot(231),plt.imshow(scharrx,'brg'),plt.title('scharrx') 
plt.subplot(232),plt.imshow(scharry,'brg'),plt.title('scharry') 
plt.subplot(233),plt.imshow(scharrxy,'brg'),plt.title('scharrxy') 
```

```python

#图像梯度-laplacian算子
'''
[0 1  0
1 -4  1
0  1  0]
'''

laplacianx=cv2.Laplacian(img_1,cv2.CV_64F)#soblex内含有负数，要绝对值一下
laplacianx = cv2.convertScaleAbs(laplacianx)
laplaciany=cv2.Laplacian(img_1,cv2.CV_64F)
laplaciany = cv2.convertScaleAbs(laplaciany)
laplacianxy=cv2.addWeighted(laplacianx,0.5,laplaciany,0.5,0)#用权值相加效果好一些
laplacianxy = cv2.convertScaleAbs(laplacianxy)

res=np.hstack((soblexy,scharrxy,laplacianxy))
cv_show('soblex',res)
plt.figure(figsize=(20,20))
plt.subplot(231),plt.imshow(soblexy,'brg'),plt.title('soblexy') 
plt.subplot(232),plt.imshow(scharrxy,'brg'),plt.title('scharrxy') 
plt.subplot(233),plt.imshow(laplacianxy,'brg'),plt.title('laplacianxy') 
```
