```python
from ast import increment_lineno
import cv2
from cv2 import cvtColor#opencv读取格式BGR
import matplotlib
import matplotlib.pyplot as plt
import numpy as np#数值计算工具包

def cv_show(name,img):
    cv2.imshow(name,img)
    cv2.waitKey(1000)#waitKey应该是会影响其他进程，一直等待任意按键
    cv2.destroyAllWindows()
```

```python
#视频播放

vc=cv2.VideoCapture(r'C:\Users\wangkai\Desktop\1.mp4')
if vc.isOpened():
    open,frame=vc.read()
else:
    open=False
while open:
    ret,frame=vc.read()
    if frame is None:
        break
    if ret == True:
        gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        cv2.imshow('result',gray)
        if cv2.waitKey(10)&0xFF==27:
            break
vc.release()
cv2.destroyAllWindows()

```
