```python
import threading
'''
给你一个类：

public class Foo {
public void first() { print("first"); }
public void second() { print("second"); }
public void third() { print("third"); }
}
三个不同的线程 A、B、C 将会共用一个 Foo实例。

线程 A 将会调用 first() 方法
线程 B 将会调用 second() 方法
线程 C 将会调用 third() 方法
请设计修改程序，以确保 second() 方法在 first() 方法之后被执行third() 方法在 second() 方法之后被执行。

提示：

尽管输入中的数字似乎暗示了顺序，但是我们并不保证线程在操作系统中的调度顺序。
你看到的输入格式主要是为了确保测试的全面性。

'''

#法一：Lock锁对象法：

class Foo:
    def __init__(self):
        self.l1 = threading.Lock()
        self.l1.acquire()
        self.l2 = threading.Lock()
        print('t1')
        self.l2.acquire()
        print('t2')
        
    def first(self): 
        print('First\n')
        self.l1.release()

    def second(self):
        print('t3')
        self.l1.acquire()
        print('Second\n')
        self.l2.release()

    def third(self) :
        print('t4')
        self.l2.acquire()
        print('Third\n')



def job1():
    f.first()
    
def job2():
    f.second()

def job3():
    f.third()

if __name__ == '__main__':
    
    f=Foo()

    t1 = threading.Thread(target=job1)
    t3 = threading.Thread(target=job3)
    t2 = threading.Thread(target=job2)
    
    t1.start()
    t3.start() 
    t2.start()
```

# 线程 A 将会调用 foo()方法，而线程 B 将会调用 bar()方法,请设计修改程序，以确保 "foobar" 被输出 n 次。

```python
  
import threading
'''
两个不同的线程将会共用一个 FooBar实例:

线程 A 将会调用 foo()方法，而
线程 B 将会调用 bar()方法
请设计修改程序，以确保 "foobar" 被输出 n 次。

输入:n = 1
输出："foobar"
解释：这里有两个线程被异步启动。其中一个调用 foo() 方法, 另一个调用 bar() 方法，"foobar" 将被输出一次。

输入:n = 2
输出："foobarfoobar"
解释："foobar" 将被输出两次。
'''

class FooBar(object):
    def __init__(self, n):
        self.n = n
        self.l1 = threading.Lock()
        self.l1.acquire()
        self.l2 = threading.Lock()


        
    def foo(self):
        """
        :type printFoo: method
        :rtype: void
        """
        for i in range(self.n):
            
                # printFoo() outputs "foo". Do not change or remove this line.
                self.l2.acquire()
                print('foo')
                self.l1.release()


    def bar(self):
        """
        :type printBar: method
        :rtype: void
        """
        for i in range(self.n):
            
                # printBar() outputs "bar". Do not change or remove this line.
                self.l1.acquire()
                print('bar')
                self.l2.release()
def job1():
    f.foo()

def job2():
    f.bar()

if __name__ == '__main__':
    
    f=FooBar(3)

    t1 = threading.Thread(target=job1)
    t2 = threading.Thread(target=job2)

    
    t1.start()
    t2.start() 
```

# 线程 A:调用 zero() ，只输出 0,线程 B:调用 even() ，只输出偶数,线程 C:调用 odd() ，只输出奇数,修改给出的类，以输出序列 "010203040506..." ，其中序列的长度必须为 2n 。

```python

import threading

'''
现有函数 printNumber 可以用一个整数参数调用，并输出该整数到控制台。

例如，调用 printNumber(7) 将会输出 7 到控制台。
给你类 ZeroEvenOdd 的一个实例，该类中有三个函数:zero、even 和 odd 。ZeroEvenOdd 的相同实例将会传递给三个不同线程：

线程 A:调用 zero() ，只输出 0
线程 B:调用 even() ，只输出偶数
线程 C:调用 odd() ，只输出奇数
修改给出的类，以输出序列 "010203040506..." ，其中序列的长度必须为 2n 。

实现 ZeroEvenOdd 类：

ZeroEvenOdd(int n) 用数字 n 初始化对象，表示需要输出的数。
void zero(printNumber) 调用 printNumber 以输出一个 0 。
void even(printNumber) 调用printNumber 以输出偶数。
void odd(printNumber) 调用 printNumber 以输出奇数。

'''
import threading

class ZeroEvenOdd(object):
    def __init__(self, n):
        self.n = n
        self.t=n+1
        self.r =n
        self.l1 = threading.Lock()

        self.l2 = threading.Lock()
        self.l2.acquire()
        self.l3 = threading.Lock()
        self.l3.acquire()
        
	# printNumber(x) outputs "x", where x is an integer.
    def zero(self, printNumber):
        """
        :type printNumber: method
        :rtype: void
        """
        while True:
            self.l1.acquire()
            if self.r==0:
                self.l2.release()
                self.l3.release()
                break
            printNumber(0)
            self.n=self.t-self.r
            if self.n%2==0:
                self.l2.release()
            else:
                self.l3.release()
        
    def even(self, printNumber):
        """
        :type printNumber: method
        :rtype: void
        """
        for i in range(self.r//2):
            self.l2.acquire()       
            printNumber(self.n)
            self.r=self.r-1
            self.l1.release()
        
        
    def odd(self, printNumber):
        """
        :type printNumber: method
        :rtype: void
        """
        for i in range(self.r//2+self.r%2):
            self.l3.acquire()
            printNumber(self.n)
            self.r=self.r-1
            self.l1.release()

def printNumber(n):
    print(n)

def job1():
    f.zero(printNumber)

def job2():
    f.even(printNumber)

def job3():
    f.odd(printNumber)
        
if __name__ == '__main__':
    
    f=ZeroEvenOdd(8)

    t1 = threading.Thread(target=job1)
    t2 = threading.Thread(target=job2)
    t3 = threading.Thread(target=job3)
    
    t1.start()
    t2.start() 
    t3.start()
```

```python

'''
现在有两种线程，氧 oxygen 和氢 hydrogen，你的目标是组织这两种线程来产生水分子。

存在一个屏障（barrier）使得每个线程必须等候直到一个完整水分子能够被产生出来。

氢和氧线程会被分别给予 releaseHydrogen 和 releaseOxygen 方法来允许它们突破屏障。

这些线程应该三三成组突破屏障并能立即组合产生一个水分子。

你必须保证产生一个水分子所需线程的结合必须发生在下一个水分子产生之前。

换句话说:

如果一个氧线程到达屏障时没有氢线程到达，它必须等候直到两个氢线程到达。
如果一个氢线程到达屏障时没有其它线程到达，它必须等候直到一个氧线程和另一个氢线程到达。
书写满足这些限制条件的氢、氧线程同步代码。

'''
class H2O(object):
    def __init__(self):
        pass
        self.num_h=0
        self.Semaphore_o = threading.Semaphore(0)
        self.Semaphore_h = threading.Semaphore(0)
    

    def hydrogen(self):
        """
        :type releaseHydrogen: method
        :rtype: void
        """
        if self.num_h==1:
            # releaseHydrogen() outputs "H". Do not change or remove this line.
            print('H')
            self.Semaphore_o.release()
            self.Semaphore_h.acquire()
            
        else:
            self.num_h+=1
            #print(self.num_h)
            # releaseHydrogen() outputs "H". Do not change or remove this line.
            print('H')

    def oxygen(self):
        """
        :type releaseOxygen: method
        :rtype: void
        """
        self.Semaphore_o.acquire()
        # releaseOxygen() outputs "O". Do not change or remove this line.
        print('O')
        self.num_h=0
        self.Semaphore_h.release()


def job1():
    for i in range(20):
        f.hydrogen()

def job2():
    for i in range(10):
        f.oxygen()  
if __name__ == '__main__':
    
    f=H2O()

    t1 = threading.Thread(target=job1)
    t2 = threading.Thread(target=job2)

    
    t1.start()
    t2.start() 
```

# 编写一个可以从 1 到 n 输出代表这个数字的字符串的程序

```python
import threading
'''
编写一个可以从 1 到 n 输出代表这个数字的字符串的程序，但是：

如果这个数字可以被 3 整除，输出 "fizz"。
如果这个数字可以被 5 整除，输出 "buzz"。
如果这个数字可以同时被 3 和 5 整除，输出 "fizzbuzz"。
例如，当 n = 15，输出： 1, 2, fizz, 4, buzz, fizz, 7, 8, fizz, buzz, 11, fizz, 13, 14, fizzbuzz。


'''

class FizzBuzz:
    def __init__(self, n):
        self.n = n
        self.lock_fizz = threading.Semaphore(0)
        self.lock_buzz = threading.Semaphore(0)
        self.lock_fizz_buzz = threading.Semaphore(0)
        self.lock_number = threading.Semaphore(1)

    def fizz(self) :
        for i in range(1, self.n + 1):
            if i % 3 == 0 and not i % 5 == 0:
                self.lock_fizz.acquire()
                print('Fizz')
                self.lock_number.release()

    def buzz(self) :
        for i in range(1, self.n + 1):
            if i % 5 == 0 and not i % 3 == 0:
                self.lock_buzz.acquire()
                print('Buzz')
                self.lock_number.release()

    def fizzbuzz(self) :
        for i in range(1, self.n + 1):
            if i % 3 == 0 and i % 5 == 0:
                self.lock_fizz_buzz.acquire()
                print('FizzBuzz')
                self.lock_number.release()

    def number(self):
        for i in range(1, self.n + 1):
            self.lock_number.acquire()
            if i % 3 == 0 and i % 5 == 0:
                self.lock_fizz_buzz.release()
            elif i % 3 == 0:
                self.lock_fizz.release()
            elif i % 5 == 0:
                self.lock_buzz.release()
            else:
                print(i)
                self.lock_number.release()
                
def job1():
        f.fizz()

def job2():
        f.buzz()

def job3():
        f.fizzbuzz()
        
def job4():
        f.number()
if __name__ == '__main__':
    
    f=FizzBuzz(30)

    t1 = threading.Thread(target=job1)
    t2 = threading.Thread(target=job2)
    t3 = threading.Thread(target=job3)
    t4 = threading.Thread(target=job4)
    
    t1.start()
    t2.start()
    t3.start()
    t4.start() 
```
